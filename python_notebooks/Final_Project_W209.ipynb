```python
data = pd.read_csv('migration_data_2010_2020.csv')
```

```python
data.head()
```

```python
#Split FULL2_NAME column by ","
split_name2 = data['FULL2_NAME'].str.split(',',expand=True)
```

```python
split_name2
```

```python
# Didn't work.
#Created a new column "State2" and tried to put values from second and third column, depending if there was a value in the third column 
split_name2.loc[split_name2[2] == None, 'State2'] = split_name2[1]
split_name2.loc[split_name2[2] != None, 'State2'] = split_name2[2]
print(split_name2)
```

```python
split_name2.columns = ['FIRST','SECOND','THIRD','State2']
```

```python
split_name2
```

```python
# Filled all the empty values in the "STATE2" column with values from the "SECOND" column
split_name2.State2.fillna(split_name2.SECOND, inplace=True)
#Rename columns
split_name2.columns = 'FIRST SECOND THIRD STATE2'.split()
```

```python
split_name2
```

```python
#Created a function that would put each field value in a list if it contained the word "County"
def county_identifier(search_string):
    county_list = []
    
    if "County" in search_string: 
        county_list.append(search_string)
    else: 
        pass
    return county_list
```

```python
#Simultaneously calling the function and applying the values from the list into a new column "county_list2"
split_name2['county_list2'] = split_name2.FIRST.apply(county_identifier)
```

```python
split_name2
```

```python
#The previous function put all the values in the new column in a list form (ex. [Davidson County]). I'm creating a new function to undo that and "stringifying" the values
def stringifier(county_list2):
    string = ', '.join(county_list2)
    return string
```

```python
# Calling the function and applying the new "string" values into a new column 
split_name2['county2'] = split_name2.county_list2.apply(stringifier)
```

```python
split_name2
```

```python
#Didn't work. Apparently, pandas didn't recognize empty fields in the "county2" column. 
split_name2.county2.fillna(split_name2.SECOND, inplace=True)
```

```python
split_name2
```

```python
#So I found another method. The r'/s*$' means white space and empty space. So I'm replacing the empty space with Nan values. 
#And then filling the Nan values with values from the "SECOND" column 
split_name2.county2 = split_name2.county2.replace(r'^\s*$', np.nan, regex=True).fillna(split_name2.SECOND)
```

```python
split_name2
```

```python
#Grabbing the "county2" and "STATE2" column
county2 = split_name2['county2']
state2 = split_name2['STATE2']
```

```python
#Joining the columns into the main dataset
data_final1 = data.join(county2)
data_final2 = data_final1.join(state2)
```

```python
#Removing the word "County" in the "county2" column
banned = ['County']
f = lambda x: ' '.join([item for item in x.split() if item not in banned])
data_final2["county2"] = data_final2["county2"].apply(f)
```
