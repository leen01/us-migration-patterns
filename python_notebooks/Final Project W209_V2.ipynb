## Fixing Alaska

```python
#Replacing "Alaska" values in "county2" column with value from FULL_NAME2 column
data['county2'] = data.apply(lambda x: x['county2'].replace('Alaska',x['FULL2_NAME']), axis=1)
```

```python
#Verifying
data.loc[data['county2'] == 'Anchorage Municipality, Alaska']
```

```python
#Deleting field string ", Alaska" in "county2" column  
data['county2'] = data['county2'].str.replace(', Alaska', '')
```

```python
#Verifying
data.loc[data['county2'] == 'Anchorage Municipality']
```
